```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(2,10,11)
```

```python
f = x+3
```

```python
y = f + np.random.normal(0,1.1,11)
```

```python
plt.scatter(x,y)
```

```python
w,b = 0,0
yp = w*x + b
plt.scatter(x,y)
plt.plot(x, yp)
for k in range(50):
    w = w + np.mean(2*x*(y-yp))*0.01
    b = b + np.mean(2*(y-yp))*0.01
    yp = w*x + b
    plt.plot(x,yp)
```

```python
# Pytorch with Regression
```

```python
X,Y = make_regression(n_features=4)
```

```python
X = X.round(2)
Y = Y.round(2)
```

```python
xtor = torch.FloatTensor(X)
ytor = torch.FloatTensor(Y)
```

```python
mod = nn.Sequential(
    nn.Linear(4, 10),
    nn.ReLU(),
    nn.Linear(10,4),
    nn.ReLU(),
    nn.Linear(4,1)
)
```

```python
losfn = nn.MSELoss()
opt = Adam(mod.parameters(), lr=0.1)
```

```python
for k in range(10):
    opt.zero_grad()
    Yp = mod(xtor)
    loss = losfn(Yp, ytor)
    loss.backward()
    opt.step()
    print(loss.item())
```
